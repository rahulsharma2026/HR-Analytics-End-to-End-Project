```python
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df['Attrition'].value_counts()
```

```python
attrition_rate = (df['Attrition'] == 'Yes').mean() * 100

print(f"Attrition Rate: {attrition_rate:.1f}%")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Attrition', data=df)

plt.title("Employee Attrition Count")
plt.show()
```

```python
pd.crosstab(df['OverTime'], df['Attrition'])
```

```python
pd.crosstab(df['OverTime'], df['Attrition']).plot(
    kind='bar',
    figsize=(6,4)
)

plt.title("Overtime vs Attrition")
plt.ylabel("Employees")
plt.show()
```

```python
dept_attrition = pd.crosstab(
    df['Department'],
    df['Attrition']
)

dept_attrition
```

```python
dept_attrition.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Department Wise Attrition")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='Age',
    bins=20
)

plt.title("Employee Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='MonthlyIncome',
    bins=30
)

plt.title("Monthly Income Distribution")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64'])

plt.figure(figsize=(14,10))

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Attrition',
    y='MonthlyIncome',
    data=df
)

plt.title("Salary vs Attrition")
plt.show()
```

```python
Key Findings

1. Overall Attrition Rate = 16.1%

2. Sales Department has highest attrition.

3. Overtime employees leave more frequently.

4. Young employees (18-25) show highest attrition.

5. Lower salary employees have higher attrition.

6. Work-life balance impacts retention.

7. Research Scientists and Sales Representatives are high-risk roles.
```
